```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img = cv2.imread("C:/Users/Karishma Sajjad/Downloads/mamacoco.jpg")
Cimage = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title("mamacoco")
plt.imshow(Cimage)
```

```python
from PIL import ImageOps
```

```python
array = np.array(Cimage)
width, height, C = array.shape
print('width', width ,'height', height,'Channels',C)
```

```python
array_flip = np.zeros((width, height, C), dtype=np.uint8)
```

```python
for i,row in enumerate(array):
    array_flip[width - 1 - i, :, :] = row
```

```python
plt.title("Flip Image of mamacoco")
plt.imshow(array_flip)
```

```python
Cimage = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
f, axarr = plt.subplots(1,2)
axarr[0].imshow(Cimage)
axarr[1].imshow(array_flip)
```
